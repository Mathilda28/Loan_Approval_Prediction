## Conceptual Problems

1. Apa yang dimaksud dengan criterion pada Decision Tree ? Jelaskan criterion yang kalian pakai dalam kasus ini !
Sebenarnya saya menggunakan DTC saat pemodelan tapi bukan best model dari data ini jadi yang dimaksud dengan criterion pada dtc adalah parameter yang digunakan untuk mengukur kualitas pemisahan pada setiap node dalam pembangunan pohon keputusan atau menentukan metrik yang digunakan untuk memilih fitur terbaik yang akan digunakan dalam membagi node. yang digunkan pada case ini menggunakan Gini Impurity sebagai criterion default untuk mengukur sejauh mana sebuah node berisi sampel-sampel dari kelas yang berbeda. Semakin kecil nilai Gini Impurity, semakin baik pemisahan pada node tersebut. Dalam scikit-learn, jika tidak ditentukan secara eksplisit, Gini Impurity merupakan nilai default yang digunakan.

2. Jelaskan apa yang dimaksud dengan pruning pada Tree-based model (alasan, definisi, jenis, dll) !
Jadi pada pruning itu merupakan proses pengurangan atau pemangkasan pohon keputusan (decision tree) dengan tujuan mengurangi kompleksitas dan meningkatkan generalisasi model. Tujuan utama pruning adalah menghindari overfitting, di mana model terlalu menyesuaikan dengan data pelatihan dan tidak mampu menggeneralisasi dengan baik pada data yang belum pernah dilihat sebelumnya. Untuk definisi Pruning adalah teknik dalam model berbasis pohon yang menghilangkan cabang-cabang (subtree) yang tidak signifikan dari pohon untuk meningkatkan kinerja model dan mencegah overfitting. Jenis pruning terdapat 2 yaitu pre_pruning dan post_pruning dimana saat pro_pruning dilakukan sebelum pemodelan selama proses pembangunan pohon keputusan, di mana pemilihan saat cabang mana yang harus dipecah atau penghentian pembangunan pohon dilakukan berdasarkan kriteria yang ditentukan sebelumnya. Beberapa teknik pre-pruning yang umum digunakan termasuk: Maximum Depth: Pembangunan pohon dihentikan setelah mencapai tingkat kedalaman tertentu. Minimum Sample Split: Pembangunan pohon dihentikan jika jumlah sampel di node saat ini tidak memenuhi ambang batas yang ditentukan. Minimum Information Gain: Pembangunan pohon dihentikan jika peningkatan informasi yang diperoleh dari pemisahan node saat ini kurang dari ambang batas yang ditentukan. Maximum Leaf Nodes: Pembangunan pohon dihentikan setelah jumlah daun mencapai ambang batas yang ditentukan. Dan untuk Post-pruning dilakukan setelah pohon keputusan dibangun sepenuhnya. Pada tahap ini, pemangkasan dilakukan dengan menghilangkan cabang-cabang yang dianggap tidak signifikan atau redundan. Beberapa teknik post-pruning yang umum digunakan termasuk: Cost Complexity Pruning (CCP): Menggunakan parameter kompleksitas dan akurasi untuk menentukan node-node yang harus dipangkas. Pruning dilakukan dengan memilih subpohon yang memiliki cost complexity terendah berdasarkan metrik evaluasi yang diberikan. Reduced Error Pruning: Melakukan validasi silang (cross-validation) untuk mengevaluasi kinerja model pada setiap tahap pemangkasan. Jika pemangkasan meningkatkan atau tidak berdampak signifikan pada kinerja model pada data validasi, maka pemangkasan dilakukan.

3. Bagaimana cara memilih K yang optimal pada KNN ?
Pada KKN untuk memilih K yang optimal untuk mencoba beberapa nilai K yang berbeda dan melakukan evaluasi yang cermat menggunakan teknik validasi silang untuk memilih nilai K yang memberikan performa model yang paling baik dan stabil. pertama kita membagi data set atau folds yang seimbang, kita membuat inisialisasi parameter dan evaluasi performa model menggunakan metrik evaluasi yang relevan (misalnya, akurasi, precision, recall, F1-score) pada set validasi. Lalu hitung skor rata-rata dan deviasi standar dari metrik evaluasi yang diperoleh pada setiap lipatan untuk setiap nilai K. Gunakan skor ini untuk membandingkan performa KNN dengan nilai K yang berbeda. Setelah memilih K yang optimal, latih ulang model KNN menggunakan seluruh dataset pelatihan dan uji performa model pada dataset uji yang belum pernah dilihat sebelumnya menggunakan K yang terpilih.

4. Jelaskan apa yang dimaksud dengan Cross Validation !
Untuk memperkirakan performa suatu model secara objektif pada data yang belum pernah dilihat sebelumnya. dimana tujuan dari cross-validation adalah untuk mengukur sejauh mana model yang telah dilatih dapat melakukan generalisasi pada data baru.

5. Apa yang dimaksud dengan metrics-metrics berikut : Accuracy, Precision, Recall, F1 Score, dan kapan waktu yang tepat untuk menggunakannya ?
- Accuracy adalah mengukur sejauh mana model secara keseluruhan dapat mengklasifikasikan data dengan benar. Akurasi dapat digunakan ketika kelas-kelas dalam dataset memiliki distribusi yang seimbang dan tidak ada perbedaan signifikan dalam dampak dari kesalahan positif palsu (False Positive) dan kesalahan negatif palsu (False Negative).
- Precision adalah (presisi) mengukur sejauh mana model benar dalam mengklasifikasikan sampel-sampel positif. Dan penggunaanya ketika fokus pada pengurangan kesalahan positif palsu (False Positive) lebih penting daripada menghindari kesalahan negatif palsu (False Negative). Misalnya, dalam kasus deteksi spam email, penting untuk meminimalkan kemungkinan email yang bukan spam salah diklasifikasikan sebagai spam.
- Recall adalah mengukur sejauh mana model dapat mengidentifikasi dan mengklasifikasikan dengan benar semua sampel positif yang ada.
- F1 Score digunakan ketika keseimbangan antara precision dan recall penting. Misalnya, ketika terdapat ketidakseimbangan kelas atau kesalahan positif palsu dan kesalahan negatif palsu memiliki konsekuensi yang sama.

````python
#Menghitung Variance Inflation Factor (VIF)
def calc_vif(X):

    # Calculating VIF
    vif = pd.DataFrame()
    vif["variables"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

    return(vif)
````

VIF digunakan untuk mendeteksi masalah multikolinieritas, yaitu adanya korelasi tinggi antara variabel independen.

## Preprocess


Pada proses dataset ini kita tidak menggunakan seperti handling outlier atau scaling karena base dari dataset ini merupakan dataset kategorikal tapi sesuai dengan aturan harus di ubah menjadi numerik. Saat ingin memproses dataset ini otomatis harus menggunakan pengerjaan layaknya data kategori makanya saat preprocess kita menggunakan imputer untuk handling dan ordinal encoder untuk mengubah variabel kategorikal yang memiliki tingkatan atau urutan menjadi representasi numerik dengan penggunaan pipeline untuk menggabungkan beberapa langkah pemrosesan data secara berurutan.

````python
#Pipeline Categorikal
cat_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('ordinal_enc', OrdinalEncoder(categories=[[-2, -1, 0, 1, 2, 3, 4, 5, 6, 7, 8],
                                               [-2, -1, 0, 1, 2, 3, 4, 5, 6, 7],
                                               [-2, -1, 0, 1, 2, 3, 4, 5, 6, 7],
                                               [-2, -1, 0, 1, 2, 3, 4, 5, 6, 7, 8],
                                               [-2, -1, 0, 1, 2, 3, 4, 5, 6, 7],
                                               [-2, -1, 0, 1, 2, 3, 4, 5, 6, 7]]))

])

cat_tr = cat_pipeline.fit_transform(X_train[['pay_0','pay_2', 'pay_3', 'pay_4', 'pay_5', 'pay_6']])
cat_tr
````

Pipeline yang kita gunakan disini hanya untuk pipeline categorikal karena data asli pada dataset kita ini adalah data categorikal. Meskipun tidak ada missing value tapi kita tetap membuat imputer untuk melihat sewaktu-waktu ada perubahan yang terjadi dengan menginput nilainya dengan modus dan penggunaan ordinal encoder karena data data yang berurut itu ordinal encoder dengan categori yang sebelumnya kita menggunakan unique untuk melihat data angkanya.

````python
# transform Test model
test_tr = cat_pipeline.transform(X_test)
test_tr
````

## Cross Validation

````python
#Cross-validation Logreg
logreg_pipe = Pipeline([
        ('preprocess', cat_pipeline),
        ('model', LogisticRegression(random_state=42))
    ])
scores_logreg = cross_val_score(logreg_pipe, X_train, y_train,
                         scoring="recall", cv=10)

print(scores_logreg)
print(scores_logreg.mean())
````

Dengan menggunakan cross-validation, kita dapat menghindari bias yang mungkin muncul akibat pemilihan dataset training dan testing yang spesifik. Hasil cross-validation memberikan gambaran yang lebih objektif tentang performa model Logistic Regression secara umum dimana rata-rata dari logreg itu 0,23. Penggunaan recall sendiri untuk melihat bagaimana menanggulangi pengaruh dari False positif (1-0) di tidak bayar tapi terbaca bayar. Hal tersebut sangat berpengaruh buruk pada perusahaan yang membuat nilai kerugian yang cukup tinggi jadi pada cross validation kita dapat melihat atau mengetahu model terbaik dari penggunaan recall.

````python
#Cross-validation SVM
svm_pipe = Pipeline([
        ('preprocess', cat_pipeline),
        ('model', SVC(random_state=42))
    ])
scores_svm = cross_val_score(svm_pipe, X_train, y_train,
                         scoring="recall", cv=10)

print(scores_svm)
print(scores_svm.mean())
````

Pada cross validation svm menghasilkan rata-rata sebesar 0,4 dalam perbandingan dengan model Logistic Regression yang sebelumnya, terlihat bahwa model SVM memiliki nilai recall yang lebih tinggi. Hal ini menunjukkan bahwa model SVM cenderung lebih baik dalam mengidentifikasi instance yang sebenarnya positif.

````python
#Cross-validation DecisionTreeClassifier
dtc_pipe = Pipeline([
        ('preprocess', cat_pipeline),
        ('model', DecisionTreeClassifier(random_state=42))
    ])
scores_dtc = cross_val_score(dtc_pipe, X_train, y_train,
                         scoring="recall", cv=10)

print(scores_dtc)
print(scores_dtc.mean())
````

Pada cross validation dengan pemodelan decision tree atau tree best model ini nilai rata-rata dari nilai recall tersebut adalah 0.372. jadi hasilnya model DTC menunjukkan performa yang lebih rendah dalam hal recall dibandingkan dengan model Logistic Regression dan SVM.

````python
#Cross-validation RandomForestClassifier
rnd_pipe = Pipeline([
        ('preprocess', cat_pipeline),
        ('model', RandomForestClassifier(random_state=42))
    ])
scores_rnd = cross_val_score(rnd_pipe, X_train, y_train,
                         scoring="recall", cv=10)

print(scores_rnd)
print(scores_rnd.mean())
````

Jadi pada cross validation dengan menggunakan randomforest menunjukkan bahwa model RFC memiliki nilai recall yang bervariasi pada setiap fold cross-validation. Rata-rata recall dari seluruh fold adalah sekitar 0.392. Recall mengukur sejauh mana model mampu mengidentifikasi dengan benar instance yang sebenarnya positif. Dalam perbandingan dengan model-model sebelumnya, terlihat bahwa model RFC memiliki nilai recall yang lebih tinggi daripada model DTC, tetapi sedikit lebih rendah daripada model SVM.

````python
#Cross-validation KNeighborsClassifier
knn_pipe = Pipeline([
        ('preprocess', cat_pipeline),
        ('model', KNeighborsClassifier())
    ])
scores_knn = cross_val_score(knn_pipe, X_train, y_train,
                         scoring="recall", cv=10)

print(scores_knn)
print(scores_knn.mean())
````

Berdasarkan hasil cross validation dengan penggunaan KNN model menunjukkan bahwa model KNN memiliki nilai recall yang bervariasi pada setiap fold cross-validation. Rata-rata recall dari seluruh fold adalah sekitar 0.394. Recall mengukur sejauh mana model mampu mengidentifikasi dengan benar instance yang sebenarnya positif. Dalam perbandingan dengan model-model sebelumnya, terlihat bahwa model KNN memiliki nilai recall yang cukup konsisten dengan model SVM, dan sedikit lebih baik daripada model Logistic Regression dan Decision Tree Classifier.


````python
#Cross-validation GaussianNB
nb_pipe = Pipeline([
        ('preprocess', cat_pipeline),
        ('model', GaussianNB())
    ])
scores_nb = cross_val_score(nb_pipe, X_train, y_train,
                         scoring="recall", cv=10)

print(scores_nb)
print(scores_nb.mean())
````

Rata-rata recall dari seluruh fold adalah sekitar 0.445. Recall mengukur sejauh mana model mampu mengidentifikasi dengan benar instance yang sebenarnya positif. Dalam perbandingan dengan model-model sebelumnya, terlihat bahwa model Gaussian NB memiliki nilai recall yang lebih tinggi daripada model DTC, KNN, Logistic Regression, dan SVM. Namun, model Gaussian NB memiliki performa yang sedikit lebih rendah daripada model RFC. Jadi, secara umum, model Gaussian NB menunjukkan performa yang cukup baik dalam hal recall, dan memiliki performa yang lebih baik daripada beberapa model lain yang telah diuji.

````python
#Cross-validation GradientBoostingClassifier
xgboost_pipe = Pipeline([
        ('preprocess', cat_pipeline),
        ('model', XGBClassifier(random_state=55))
    ])
scores_xgboost = cross_val_score(xgboost_pipe, X_train, y_train,
                         scoring="recall", cv=10)

print(scores_xgboost)
print(scores_xgboost.mean())
````

Hal ini menunjukkan bahwa model XGBoost memiliki nilai recall yang bervariasi pada setiap fold cross-validation. Rata-rata recall dari seluruh fold adalah sekitar 0.390. Recall mengukur sejauh mana model mampu mengidentifikasi dengan benar instance yang sebenarnya positif. Dalam perbandingan dengan model-model sebelumnya, terlihat bahwa model XGBoost memiliki nilai recall yang lebih rendah daripada model Gaussian NB, RFC, dan KNN. Namun, model XGBoost memiliki performa yang sedikit lebih baik daripada model Logistic Regression, Decision Tree Classifier, dan SVM.

````python
nama_model = []
cv_scores=0
for cv,name in zip([scores_logreg, scores_svm, scores_dtc, scores_rnd, scores_knn, scores_nb, scores_xgboost],
    ['log_mod','svm_mod', 'dtc_mod', 'rnd_mod', 'knn_mod', 'nb_mod', 'xgboost_mod']):
    print(name)
    print('All - Cross Validation  : ', cv)
    print('Mean - Cross Validation : ', cv.mean())
    print('-'*50)
    if cv.mean()>cv_scores:
        cv_scores = cv.mean()
        nama_model = name
    else:
        pass
print('best model:', nama_model)
print('cross-val mean:', cv_scores)
````
